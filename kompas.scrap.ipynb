```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

```python
url = "https://tirto.id/"
response = requests.get(url)
conn = sqlite3.connect('scrap.news.db')
```

```python
soup = BeautifulSoup(response.text, "lxml")
links = soup.find_all("a")
data = []
```

```python
for link in links:
    data_text = link.text.replace("\n","")
    data_href = link.get("href")
    data_img = None
    if link.find("img") != None:
        data_img = link.find("img").get("alt")

    if data_text == "":
        data_text = data_img
        
    row = [data_text,data_href]
    data.append(row)

df = pd.DataFrame(data,columns=['title','href'])
df.to_sql('TIRTO', conn, if_exists='replace', index = False)
```
